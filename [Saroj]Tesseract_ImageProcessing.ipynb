```
#Importing necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
import pytesseract
try:
  from PIL import Image
except ImportError:
  import Image
```

```
#Recognizing the text in image using pytesseract

filename = 'nep.png'
img = Image.open(filename)
custom_config = '--oem 3 --psm 6' #oem is engine mode and psm is page segmentation mode
extractedText = pytesseract.image_to_string(img, lang='nep',config = custom_config)
print(extractedText)
```

```
filename = 'eng.png'
img = Image.open(filename)
custom_config = '--oem 3 --psm 6' 
extractedText = pytesseract.image_to_string(img, lang='eng',config = custom_config)
print(extractedText)
```

```
import cv2
img = cv2.imread('xyz.png')
```

```
#For grayscale
def get_grayscale(image):
  return cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

img_gray = get_grayscale(img)
plt.imshow(img_gray)
```

```
#Noise Removal

def remove_noise(image):
  return cv2.medianBlur(img, 9)

img_noise = remove_noise(img)
plt.imshow(img_noise)
```

```
#Dilation

def dilation(image):
  kernel = np.ones((5,5),np.uint8)
  return cv2.dilate(img, kernel, iterations=1)
img_dilate = dilation(img)
plt.imshow(img_dilate)
```

```
#Erosion

def erosion(image):
  kernel = np.ones((5,5), np.uint8)
  return cv2.erode(img,kernel,iterations=2)
img_erode = erosion(img)
plt.imshow(img_erode)
```
